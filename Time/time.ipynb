```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
```


- `timestamp`: 时间戳, 从 1970 年 1 月 1 日 0 时 0 分 0 秒到现在的秒数
- `struct_time`: 9 个元素的元组, 用来表示一个时间点
- `format time`: 格式化时间, 用来表示一个时间点
- `sleep`: 休眠, 使程序进入睡眠状态, n 秒后继续

三种时间表示方法的转换关系:



# 1 `timestamp` 时间戳
`time.time()` 返回当前时间的时间戳, 从 1970 年 1 月 1 日 0 时 0 分 0 秒到现在的秒数

```python
# 返回当前时间的时间戳（1970纪元后经过的浮点秒数）
start_time = time.time()


s = ""
for i in range(100000):
    s += str(i)

end_time = time.time()
print("程序运行时间", end_time - start_time, "s")
```

# 2 `struct_time` 时间元组
- `time.localtime()` 返回当前时间的时间元组
- `time.mktime()` 将时间元组转换为时间戳

```python
now = time.time()

time_tupple = time.localtime(now)

print(time_tupple.tm_year, "年", end=" ")
print(time_tupple.tm_mon, "月", end=" ")
print(time_tupple.tm_mday, "日", end=" ")
print(time_tupple.tm_hour, "时", end=" ")
print(time_tupple.tm_min, "分", end=" ")
print(time_tupple.tm_sec, "秒")

# 转换成时间戳
print(time.mktime(time.strptime("2019-1-2 12:26:43", "%Y-%m-%d %H:%M:%S")))
```

# 3 `format time` 格式化时间
- `time.strftime(format, struct_time)` 将时间元组转换为格式化时间
- `time.strptime(format_time, format)` 将格式化时间转换为时间元组


format 的格式：
- %Y: year

- %m: month

- %d: day

- %H: hour

- %M: minute

- %S: second

- %w: weekday

- %j: day of the year

- %U: week of the year (Sunday is the first day of the week)

- %W: week of the year (Monday is the first day of the week)

- %A: weekday name

- %B: month name


```python
print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
print(time.strptime("2019-1-2 12:26:43", "%Y-%m-%d %H:%M:%S"))
```
