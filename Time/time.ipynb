# 2 `struct_time` 
- `time.localtime()` 
- `time.mktime()` 

```python
now = time.time()

time_tupple = time.localtime(now)

print(time_tupple.tm_year, "", end=" ")
print(time_tupple.tm_mon, "", end=" ")
print(time_tupple.tm_mday, "", end=" ")
print(time_tupple.tm_hour, "", end=" ")
print(time_tupple.tm_min, "", end=" ")
print(time_tupple.tm_sec, "")

# 
print(time.mktime(time.strptime("2019-1-2 12:26:43", "%Y-%m-%d %H:%M:%S")))
```

# 3 `format time` 
- `time.strftime(format, struct_time)` 
- `time.strptime(format_time, format)` 


format 
- %Y: year

- %m: month

- %d: day

- %H: hour

- %M: minute

- %S: second

- %w: weekday

- %j: day of the year

- %U: week of the year (Sunday is the first day of the week)

- %W: week of the year (Monday is the first day of the week)

- %A: weekday name

- %B: month name


```python
print(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
print(time.strptime("2019-1-2 12:26:43", "%Y-%m-%d %H:%M:%S"))
```
