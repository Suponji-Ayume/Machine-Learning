````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

# 1 设置字体

`plt.rcParams['font.sans-serif'] = ['SimHei']` 支持中文

其他字体选项：['SimHei', 'KaiTi', 'Lantinghei SC', 'Microsoft YaHei', 'Times New Roman', 'Arial']

`plt.rcParams[‘axes.unicode_minus’]=False` 支持负号


1. 首先删除你的缓存。找到你的缓存目录，然后删除它。你可以用下面的代码找到你的缓存目录：

   ```python
   import matplotlib as mpl
   print(mpl.get_cachedir())
   ```

2. 下载字体

- `cd ~/.fonts` 如果没有就 `mkdir ~/.fonts`

- `wget https://link.zhihu.com/\?target\=http%3A//129.204.205.246/downloads/SimHei.ttf`

- `brew install fontconfig`

3. 刷新缓存

- `fc-cache -fv`


````python
x = np.arange(-10, 10, 0.1)
y = np.sin(x)

# 设置中文字体
plt.rcParams["font.family"] = ["sans-serif"]
plt.rcParams["font.sans-serif"] = ["SimHei"]

# 支持负号
plt.rcParams["axes.unicode_minus"] = False

plt.figure(figsize=(14, 7), dpi=100)
plt.title("正弦函数图像")
plt.plot(x, y)
plt.show()
````

# 2 设置坐标轴标签

`plt.xlabel('x轴标签', fontproperties='SimHei', fontsize=20, fontweight = 'bold', color='red', rotation=0, labelpad=20, loc='center', rotation=0 )`

- fontproperties: 字体

- fontsize: 字体大小

- fontweight: 字体粗细, 可选值: 'light', 'normal', 'medium', 'semibold', 'bold', 'heavy', 'black'

- color: 字体颜色

- rotation: 旋转角度, 逆时针旋转

- labelpad: 标签与坐标轴的距离,

- loc: 标签位置, 可选值: 'left', 'center', 'right'

- rotation: 标签旋转角度, 逆时针旋转


````python
plt.xlabel(
    "x轴",
    fontsize=14,
    color="red",
    labelpad=20,
    fontweight="bold",
    loc="right",
    rotation=315,
)
plt.ylabel(
    "y轴",
    fontsize=14,
    color="red",
    labelpad=20,
    fontweight="bold",
    loc="top",
    rotation=45,
)
````

# 3 设置坐标轴刻度

`plt.xticks(ticks=None, labels=None, rotation=0, color=None, **kwargs)`

- ticks: 刻度值, 可以是一个列表, 也可以是一个数组, 是我们想要显示的刻度值所在的位置, 从 0 开始，如果传入空列表, 默认删除所有原有刻度值

- labels: 刻度标签, 可以是一个列表, 也可以是一个数组

- rotation: 刻度标签旋转角度, 逆时针旋转

- color: 刻度标签颜色


````python
times = [
    "2014/7/1",
    "2014/8/1",
    "2014/9/1",
    "2014/10/1",
    "2014/11/1",
    "2014/12/1",
    "2015/1/1",
    "2015/2/1",
    "2015/3/1",
    "2015/4/1",
    "2015/5/1",
    "2015/6/1",
    "2015/7/1",
    "2015/8/1",
    "2015/9/1",
    "2015/10/1",
    "2015/11/1",
    "2015/12/1",
]

sales = np.random.randint(100, 200, size=len(times))

plt.figure(figsize=(14, 7), dpi=100)
plt.xticks(range(0, len(times), 2), times[0 : len(times) : 2], rotation=45)
plt.plot(times, sales)
plt.show()
````

# 4 设置坐标轴范围

- `plt.xlim(left=None, right=None)`
- `plt.ylim(bottom=None, top=None)`
- `plt.axis([xmin, xmax, ymin, ymax])`

如果刻度不是数字类型，那么这里的 xmin, xmax, ymin, ymax 就是刻度的位置, 而不是刻度的值

注意与 `plt.xticks` 的顺序关系


````python
plt.xticks(range(0, len(times), 2), times[0 : len(times) : 2], rotation=45)
plt.axis([2, len(times) - 2, 0, 200])
plt.plot(times, sales)
plt.show()
````
